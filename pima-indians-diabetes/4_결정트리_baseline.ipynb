## 트리모델 전반적인 과정

### 필요 라이브러리 설치

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 셀 안에 그래프표시
%matplotlib inline
```

### 데이터불러오기

```
df = pd.read_csv('/content/diabetes.csv')
```

```
df.shape
```

```
df.head()
```

### 학습데이터와 test데이터 나누기

```
# 8:2의 비율로 나누려한다
split_count = int(df.shape[0] * 0.8)
split_count
```

```
# 슬라이싱으로 나누기
train = df[:split_count].copy()  # 명시적으로 copy()붙여준다
test = df[split_count:].copy()

print(train.shape)
print(test.shape)
```

### 학습예측에 사용할 컬럼

```
# data의 columns를 보고 나눠야한다.
print(train.columns)
print(type(train.columns))
```

```
# 마지막 정답 값을 빼고 가져온다  tolist()로 리스트로 만들기 가능
feature_names = train.columns[:-1].tolist()
feature_names
```

### 정답값과 예측해야할 값

```
label_name = train.columns[-1]  # 하나만 가지고 와서 str형태로 가져온다. 
label_name
```

### 학습, 예측데이터 만들기

```
# 학습 data
X_train = train[feature_names]
print(X_train.shape)
X_train.head()
```

```
# 학습 label
Y_train = train[label_name]
print(Y_train.shape)
Y_train.head()
```

```
# test data
X_test = test[feature_names]
print(X_test.shape)
X_test.head()
```

```
# test label
Y_test = test[label_name]
print(Y_test.shape)
Y_test.head()
```

## 학습과 예측

```
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=42)
model
```

```
model.fit(X_train,Y_train) # 학습
```

```
# 예측
Y_predict = model.predict(X_test)
Y_predict
```

## tree 알고리즘 분석

```
from sklearn.tree import plot_tree
```

```
plt.figure(figsize=(20,20))
# filled 는 색 입히는 것, 
tree = plot_tree(model,
                feature_names=feature_names,
                filled = True,
                fontsize=10,
                )
```

glucose 조건이 가장 좋은것을 확인 할 수있다.    
밑으로 내려오면서 특정 feature값들을 기준으로 나눈다.    
gini 계수가 0이 될때 까지 클래스를 나눈다.

```
# graphviz로 보면 크게나오고 그래프에 더 다양한 옵션을 줄 수 있다.
import graphviz
from sklearn.tree import export_graphviz

dot_tree = export_graphviz(model, feature_names = feature_names,
filled=True, rounded = True)
graphviz.Source(dot_tree)
```

```
# 피쳐 중요도 추출
# feature의 순서에 따라서 중요도 표시
model.feature_importances_
```

```
# 피쳐중요도 시각화
sns.barplot(x=model.feature_importances_, y=feature_names)
```

### 정확도 측정

```
# 실제값 - 예측값 을 빼주면 같은 값일때 0 이나온다
# 절대값을 씌워서 다른 값들만 샌다
diff_count = abs(Y_test - Y_predict).sum()
diff_count
```

```
# 얼마나 맞추었는지 구한다
(len(Y_test) - diff_count) / len(Y_test) * 100
```

```
# 구현된 알고리즘 가져와서 사용하기
from sklearn.metrics import accuracy_score
```

```
accuracy_score(Y_test, Y_predict)
```

```
# model의 score로 점수 계산
# 테스트 데이터의 데이터를 넣고 레이블을 넣어서 계산
model.score(X_test, Y_test)
```
